### Cell Typing by Clusters

```python
import scimap as sm
import numpy as np
import anndata as ad
```

```python
# Load Data
adata = ad.read_h5ad("/mnt/disks/ecdyer-disk1/data/MEL01-3-1.h5ad")
```

```python
# Cluster
adata = sm.tl.cluster(adata, method='leiden', resolution=0.3, use_raw=False, log=False)
```

```python
# Generate Heatmap 
sm.pl.heatmap(adata, groupBy='leiden', standardScale='column', figsize=(15,15), showPrevalence=True)
```

```python
# Save data with clusters:
adata.write('/mnt/disks/ecdyer-disk1/data/leiden_clustered_MEL01-3-1.h5ad')
```
